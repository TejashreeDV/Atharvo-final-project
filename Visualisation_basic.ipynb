```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x= np.linspace(0, 10, 100) # equal spaced
x
```

```python
y = np.sin(x)
z  = np.cos(x)
y
```

```python
plt.figure(figsize= (15, 8))
plt.scatter(x, y, marker = "*", s = 60, color = "#00A73D", label = "sin curve")
plt.plot(x, z, marker = "o", linestyle = 'dotted', c = 'm', markersize = 4, label = "cos curve")
plt.title("Sin curve", fontsize = 20)
plt.xlabel("number", fontsize = 15)
plt.ylabel("sin value", fontsize = 15)
plt.xticks([ 4, 7, 8, 0, 1, 2, 3, 9, 10, 5, 6])
plt.yticks([-1, -.5, 0, 0.5, 1])
plt.xlim([0, 2*np.pi])
plt.grid()
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(15, 8))

ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.plot(x, z, marker = "o", linestyle = 'dotted', c = 'm', markersize = 4, label = "cos curve")
ax1.set_title("Sin curve", fontsize = 20)
ax1.set_xlabel("number", fontsize = 15)
ax1.set_ylabel("sin value", fontsize = 15)

ax2.scatter(x, y, marker = "*", s = 60, color = "#00A73D", label = "sin curve")
ax2.set_title("cos curve", fontsize = 20)
ax2.set_xlabel("number", fontsize = 15)
ax2.set_ylabel("cos value", fontsize = 15)
plt.show()
```
